```python
import matplotlib.pyplot as plt
import numpy as np
import fuzzyPy as fz
```

```python
dom=[0,20]
points=[5,10,17]
fz.f_triangular(dom,points,7)

dom=[0,20]
points2=[5,7,12,15]
fz.f_trapezoid(dom,points2,7)
```

```python
x=np.linspace(dom[0],dom[1],1000)

y=[fz.f_triangular(dom,points,a) for a in x]
plt.plot(x, y)

y2=[fz.f_trapezoid(dom,points2,a) for a in x]
plt.plot(x, y2)

plt.show()
```
